## 📲 Крок 5: Завантаження APK

Завантажуємо готовий APK файл на ваш комп'ютер.

```python
from google.colab import files
import glob
import os

# Пошук APK файлів
apk_files = glob.glob('bin/*.apk')

if apk_files:
    apk_path = apk_files[0]
    print(f"✅ APK знайдено: {apk_path}")
    
    # Отримання розміру файлу
    size = os.path.getsize(apk_path) / (1024 * 1024)
    print(f"📦 Розмір: {size:.2f} MB")
    
    # Завантаження
    print("📥 Завантаження...")
    files.download(apk_path)
else:
    print("❌ APK не знайдено!")
    print("Перевірте лог збірки на наявність помилок.")
    
    # Показати вміст папки bin
    !ls -la bin/ 2>/dev/null || print("Папка bin/ не існує")
```
